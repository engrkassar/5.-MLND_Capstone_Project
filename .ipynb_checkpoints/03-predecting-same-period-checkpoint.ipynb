# Same as Last Period Prediction

In this notebook we will make the baseline for our predictions, same as the last period.

```python
%matplotlib
import matplotlib.pyplot as plt

import sys
import os

import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


import boto3
import sagemaker
from sagemaker import get_execution_role

sagemaker_session = sagemaker.Session()
role = sagemaker.get_execution_role()


# S3 bucket name
bucket = 'sagemaker-us-west-2-479862333671'
subfolder = ''
conn = boto3.client('s3')
contents = conn.list_objects(Bucket=bucket, Prefix=subfolder)['Contents']
```

```python
data_ready_location = 's3://{}/{}'.format(bucket, 'data/data_ready.csv')
```

```python
data_ready = pd.read_csv(data_ready_location)
```

```python
data_ready
```

```python
Y_pred = data_ready.loc[(data_ready['d'] > 1886) & (data_ready['d'] < 1914)]['demand'].to_frame()
```

```python
Y_pred
```

```python
Y_test = pd.read_csv('s3://{}/{}'.format(bucket, 'data/val.csv'))['demand'].to_frame()
```

```python
Y_test
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mean_squared_error(Y_test, Y_pred, squared=True)
```

```python
mean_squared_error(Y_test, Y_pred, squared=False)
```

```python
mean_absolute_error(Y_test, Y_pred)
```
